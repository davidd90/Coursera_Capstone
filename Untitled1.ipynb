```python
apartments_df = pd.read_csv('Rohdaten/2019-12-23 134738.csv',delimiter=';')
apartments_df.head()
```

```python
columns= {"geo_plz","obj_baseRent","obj_regio3","obj_street","lat","long","obj_livingSpace"}
apartments_df= apartments_df[columns]
apartments_df.dtypes
```

```python
apartments_df['lat'] = apartments_df['lat'].str.replace(",",".")
apartments_df['long'] = apartments_df['long'].str.replace(",",".")

apartments_df.head()
```

```python
apartments_df['rel_baseRent'] = apartments_df['obj_baseRent'].div(apartments_df['obj_livingSpace'], axis=0)
apartments_df.head()
```

```python
apartments_df['s-stations'] = 0
for index, row in apartments_df.iterrows():
    long = float(row['long'])
    lat = float(row['lat'])
    url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}&categoryId={}'.format(CLIENT_ID, CLIENT_SECRET, lat, long, VERSION, 500, 50,'4bf58dd8d48988d1fc931735')
    results = requests.get(url).json()
    if(len(results['response'])==0):
        continue
    number_of_stations = len(results['response']['groups'][0]['items'])
    print('There are {} s-bahn stations around.'.format(number_of_stations))
    apartments_df['s-stations'][index] = number_of_stations
    
```

```python
apartments_df['culture'] = 0
for index, row in apartments_df.iterrows():
    if index < 6:
        continue
    long = float(row['long'])
    lat = float(row['lat'])
    url = 'https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&ll={},{}&v={}&radius={}&limit={}&categoryId={}'.format(CLIENT_ID, CLIENT_SECRET, lat, long, VERSION, 500, 50,'4d4b7104d754a06370d81259')
    results = requests.get(url).json()
    if(len(results['response'])==0):
        continue
    number_of_venues = len(results['response']['groups'][0]['items'])
    print('There are {} culture venues around.'.format(number_of_venues))
    apartments_df['culture'][index] = number_of_venues
    
```

```python
apartments_df['culture'].plot.hist(bins=5, alpha=0.5)
```

```python
cut = pd.cut(apartments_df.culture, [0, 5, 10, 15, 20, 25, 30, 35, 40])

boxdf = apartments_df.groupby(cut).apply(lambda apartments_df: apartments_df.rel_baseRent.reset_index(drop=True)).unstack(0)

sns.boxplot(data=boxdf);
```

```python

address = 'Leipzig, DE'

geolocator = Nominatim(user_agent="leipzig_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of leipzig are {}, {}.'.format(latitude, longitude))
```

```python
# create map of New York using latitude and longitude values
map_leipzig = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough,rel_baseRent in zip(apartments_df['lat'], apartments_df['long'], apartments_df['obj_regio3'],apartments_df['rel_baseRent']):
    label = '{}'.format(borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=rel_baseRent/2,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_leipzig)  
    
map_leipzig
```
